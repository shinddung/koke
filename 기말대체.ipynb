```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
model = LinearRegression()

X=[[226.7],[242],[241.7],[251.3],[240.5]]
y=[[14672],[15492],[15872],[17908],[16454]]
model.fit(X,y)
```

```python
print("coef:",model.coef_)
print("intercept:",model.intercept_)
```

```python
plt.title("day/accident?")
plt.xlabel("sunny day")
plt.ylabel("sunny day accident")
plt.scatter(X,y)
plt.plot(X,model.coef_*X + model.intercept_)
```

```python
x = [[224]]
y_predict = model.predict(x)
print("x: {} -> y_predict: {}".format(x,y_predict))
```
